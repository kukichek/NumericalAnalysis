### Постановка задачи

Разработать программу численного решения СЛАУ на основе $LDL^T$-разложения. 
Матрицу системы сформировать следующим образом: 
- недиагональные элементы $a_{i, j}$, $i<j$, выбираются из чисел $0, –1, –2, –3, –4$ произвольным образом; если $i>j$, то полагается $a_{i,j}$ = $a_{j,i}$;
- $a_{i,i}$ = $-\sum\limits_{j = 1, j \neq i}^{n}a_{i,j}$ ,  $2≤i≤n$;  
- $a_{1 1}$ = $-\sum\limits_{j = 2}^{n}a_{i,j} + 10^{-k}$,  $k≥0$.

Правую часть $b$ задать умножением матрицы $A$ на вектор $x=(m, m+1, ... , n+m–1):  b=Ax$.
Для вычислений выбрать параметры: 
- $m$ – номер в списке студенческой группы; 
- $n$ – одно из чисел в пределах от $10$ до $12$; 
- $k$ – рассмотреть два случая: $k=0$, $k=$ (номер студенческой группы); элементы $a_{i,j}$ при фиксированных $i$ и $j$ в обоих случаях одни и те же (матрицы отличаются только элементом $a_{1, 1}$).

Программно реализовать (в качестве языка программирования выбрать C или C++) вычисления для рассматриваемого примера. Использовать алгоритм (5) файла «LDLt_RtR разложения», требующий хранения только нижнего треугольника матрицы.

### Листинг программы

Класс, содержащий двумерную матрицу

```c++
struct Matrix {
public:

	const int width_, height_;

	Matrix(int w, int h) : width_(w), height_(h) { // создание матрицы высоты h и шириной w
        	a = new float*[height_];

        for (int i = 0; i < height_; ++i) {
            a[i] = new float[width_];
        }
    }
    
	~Matrix() {
        for (int i = 0; i < height_; ++i) {
            delete[] a[i];
        }
        delete[] a;
    }

	float* operator[] (int i) {
        return a[i];
    }

private:
	float** a;
}; 
```

Класс LES - linear equations system, то есть система линейных уравнений. 
Size - размер матрицы, offset - начальное значение вектора решений, option - переменная k

```c++
class PrintToFile;

struct LES {
public:

	LES(int size, int offset, int option) // инициализация СЛАУ, ее размера и коэффициентов
        : coefs(size, size), constTerms(1, size), offset_(offset), option_(option) {
        generateCoefs(option);
        generateCTerms(offset);

        state = "LinearEquationsSystem";
    }

	void LDLtDecompos() { // Нахождение LDLt-разложения матрицы коэффициентов
        Matrix t(1, coefs.height_);
        for (int k = 0; k < coefs.height_ - 1; ++k) {
            for (int i = k + 1; i < coefs.height_; ++i) {
                t[i][0] = coefs[i][k];
                coefs[i][k] /= coefs[k][k];

                for (int j = k + 1; j <= i; ++j) {
                    coefs[i][j] -= t[j][0] * coefs[i][k];
                }
            }
        }

        state = "LDLtDecompos";
    }
    
	void findSolution() { // Нахождение решения СЛАУ
        findY();
        findZ();
        findX();

        state = "ApproximateSolution";
    }

	float relativeError() { // Получение относительной ошибки решения
        int maxI = 0;
        for (int i = 1; i < coefs.height_; ++i) {
            if (std::fabs(constTerms[i][0] - (i + offset_)) > std::fabs(constTerms[maxI][0] - (maxI + offset_))) {
                maxI = i;
            }
        }

        return std::fabs(constTerms[maxI][0] - (maxI + offset_)) / (coefs.height_ + offset_ - 1) * 100;
    }

private:
	Matrix coefs, constTerms;
	int option_, offset_;
	std::string state;
    
	friend class PrintToFile; // Дружественный класс для вывода СЛАУ в файл

	void generateCoefs(int option) { // Генерация коэффициентов симметрической матрицы
        for (int i = 0; i < coefs.height_; ++i) {
            coefs[i][i] = 0;
            
            for (int j = i + 1; j < coefs.height_; ++j) {
                coefs[i][j] = std::rand() % 5 - 4;
                coefs[i][i] -= coefs[i][j];
            }
            
            for (int j = 0; j < i; ++j) {
                coefs[i][j] = coefs[j][i];
                coefs[i][i] -= coefs[i][j];
            }
        }

        (option == 0) ? coefs[0][0] += 1 : coefs[0][0] += 0.1; // Прибавление 1 или 0.1 
                                                               // к первому элементу матрицы в зависимости от варианта k
    }

	void generateCTerms(int offset) { // Генерация вектора решений
        for (int i = 0; i < constTerms.height_; ++i) {
            constTerms[i][0] = 0;
            for (int j = 0; j < constTerms.height_; ++j) {
                constTerms[i][0] += (j + offset) * coefs[i][j];
            }
        }
    }

	void findY() { // Нахождение решения уравнения Ly = b
        for (int k = 0; k < coefs.height_ - 1; ++k) {
            for (int i = k + 1; i < coefs.height_; ++i) {
                constTerms[i][0] -= coefs[i][k] * constTerms[k][0];
            }
        }

        state = "Y";
    }

	void findZ() { // Нахождение решения уравнения Dz = y
        for (int i = 0; i < coefs.height_; ++i) {
            constTerms[i][0] /= coefs[i][i];
        }

        state = "Z";
    }

	void findX() { // Нахождение решения уравнения L^{t}x = z
        for (int i = coefs.height_ - 1; i >= 0; --i) {
            for (int j = i + 1; j < coefs.height_; ++j) {
                constTerms[i][0] -= coefs[j][i] * constTerms[j][0];
            }
        }

        state = "X";
    }
};
```
